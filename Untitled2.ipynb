```python
import numpy as np
import pandas as pd

np.random.seed(0)

days = 30
time = np.arange(days)

T_target = 60  # دمای هدف

warmup_time = []

for t in time:
    if t < 10:
        base = 12
    elif t < 20:
        base = 13.5
    else:
        base = 16

    # نویز واقعی 
    noise = np.random.normal(0, 0.5)

    warmup_time.append(base + noise)

warmup_time = np.array(warmup_time)

df = pd.DataFrame({
    "day": time,
    "warmup_time_min": warmup_time
})

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))

plt.plot(df["day"], df["warmup_time_min"], marker="o")

plt.title("Boiler Warmup Time Over Days")
plt.xlabel("Day")
plt.ylabel("Warmup Time (min)")

plt.grid(True)
plt.show()
```
